```python
import xarray as xr
import pandas as pd
from pathlib import Path
import glob
import numpy as np
import subprocess
import os
```

```python
start_date = pd.Timestamp('2002-08-11')
end_date = pd.Timestamp('2002-08-27')
date_range = pd.date_range(start_date, end_date, freq='30min')

base_path = '/newstor/climate_data/IMERG_data/global_30min/2002'
output_dir = "/scratch/vishald/ug1_monsoonlab/pratham/Datasets/Maharashtra_August_2002/Imerg"
```

```python
start_str = pd.Timestamp(start_date).strftime('%d%b%Y')
end_str = pd.Timestamp(end_date).strftime('%d%b%Y')

temp_filename = f"IMERG0p1_from_{start_str}_to_{end_str}_resampled_12hr_final.nc"
output_filename_1deg = f"IMERG1_from_{start_str}_to_{end_str}_resampled_12hr_final.nc"

temp_filepath= f"{output_dir}/{temp_filename}"
output_filepath_1deg=f"{output_dir}/{output_filename_1deg}"

if Path(output_dir).exists():
    print(f"File {output_dir} already exists")
else:
    # Create the directory if it doesn't exist
    Path(output_dir).mkdir(exist_ok=True)
    print(f"Created non existing folder: {output_dir}")
```

```python
def get_imerg_filename(timestamp):
    # Format: 3B-HHR.MS.MRG.3IMERG.YYYYMMDD-SHHMMSS-EHHMMSS.XXXX.V07B.HDF5.SUB.nc4
    # where XXXX is minutes since 00:00 UTC
    
    date_str = timestamp.strftime('%Y%m%d')
    start_time = timestamp.strftime('S%H%M%S')
    # End time is 30 minutes later
    end_time = (timestamp + pd.Timedelta(minutes=29, seconds=59)).strftime('E%H%M%S')
    
    # Calculate minutes since 00:00 UTC
    minutes_since_midnight = timestamp.hour * 60 + timestamp.minute
    
    filename = f"3B-HHR.MS.MRG.3IMERG.{date_str}-{start_time}-{end_time}.{minutes_since_midnight:04d}.V07B.HDF5.SUB.nc4"
    return Path(base_path) / filename
```

```python
# Create a list to store datasets
datasets = []

# Open each file in the date range
for timestamp in date_range:
    try:
        filename = get_imerg_filename(timestamp)
        if filename.exists():
            # print(filename)
            ds = xr.open_dataset(filename)
            datasets.append(ds)
    except Exception as e:
        print(f"Error opening file for {timestamp}: {e}")

# Combine all datasets along the time dimension
if datasets:
    combined_ds = xr.concat(datasets, dim='time')
    print(f"Successfully loaded data from {start_date} to {end_date}")
    print(f"Dataset shape: {combined_ds.precipitation.shape}")
else:
    print("No datasets were loaded")
del datasets
```

```python
ds_first_hour=combined_ds.resample(time='1h',label='right').mean()
```

```python
assert np.allclose(ds_first_hour.precipitation.isel(time=0),combined_ds.precipitation.isel(time=slice(0,2)).mean(dim='time'),equal_nan=True)==True,'Values not matching after one hour resampling'
```

```python
diff = ds_first_hour.precipitation.isel(time=0) - combined_ds.precipitation.isel(time=slice(0,2)).mean(dim='time')
print("\nMaximum absolute difference:")
print(np.abs(diff).max().values)
```

```python
df_twelve_hour=ds_first_hour.resample(time='12h',closed='right',label='right').sum()
```

```python
assert np.allclose(df_twelve_hour.isel(time=0).precipitation.values,ds_first_hour.isel(time=slice(0,12)).precipitation.sum(dim='time').values,equal_nan=True)==True,'Value not matching after 12hr resampling compared with one hour'
```

```python
try:
    df_twelve_hour.to_netcdf(temp_filepath)
    print(f"Successfully created file: {temp_filepath}")
except e:
    print(f"Error: {temp_filepath}")
```

## Regridding to one degree

```python
cmd = [
    "cdo",
    "remapcon,grid.txt",
    str(temp_filepath),
    str(output_filepath_1deg)
]

try:
    # Run the command with shell=False
    result = subprocess.run(
        cmd,
        shell=False,
        check=True,
        text=True,
        stderr=subprocess.PIPE,
        stdout=subprocess.PIPE
    )
    print(f"Successfully remapped to {output_filepath_1deg}")
except subprocess.CalledProcessError as e:
    print(f"Error during CDO remapping:")
    print(f"Exit code: {e.returncode}")
    print(f"Error message: {e.stderr}")
```
